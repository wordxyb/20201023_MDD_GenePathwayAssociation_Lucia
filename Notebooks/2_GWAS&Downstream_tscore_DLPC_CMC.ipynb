# Generate Pheno Table

For example:
- tscore_DLPC_CMC Cluster1vs.Rest
- There are clusters 1,2,3. Now I name 2 and 3 as 0. 

```r
colnames(data)
```

```r
table(data$tscore_DLPC_CMC)
```

- Create columns: 1vsRest, 2vsRest, 3vsRest

```r
data$tscore_DLPC_CMC_1vsRest = data$tscore_DLPC_CMC
data$tscore_DLPC_CMC_1vsRest[data$tscore_DLPC_CMC_1vsRest==2]=0
data$tscore_DLPC_CMC_1vsRest[data$tscore_DLPC_CMC_1vsRest==3]=0
```

```r
data$tscore_DLPC_CMC_2vsRest = data$tscore_DLPC_CMC
data$tscore_DLPC_CMC_2vsRest[data$tscore_DLPC_CMC_2vsRest==1]=0
data$tscore_DLPC_CMC_2vsRest[data$tscore_DLPC_CMC_2vsRest==3]=0
data$tscore_DLPC_CMC_2vsRest[data$tscore_DLPC_CMC_2vsRest==2]=1
```

```r
data$tscore_DLPC_CMC_3vsRest = data$tscore_DLPC_CMC
data$tscore_DLPC_CMC_3vsRest[data$tscore_DLPC_CMC_3vsRest==1]=0
data$tscore_DLPC_CMC_3vsRest[data$tscore_DLPC_CMC_3vsRest==2]=0
data$tscore_DLPC_CMC_3vsRest[data$tscore_DLPC_CMC_3vsRest==3]=1
```

- generate pheno table

```r
clusters = c("tscore_DLPC_CMC_1vsRest","tscore_DLPC_CMC_2vsRest","tscore_DLPC_CMC_3vsRest")
for (c in clusters){
    df = data %>% select(IID,c)
    df$IID2 = df$IID
    col_order = c("IID","IID",c)
    df = df[, col_order]
    write.table(df, file = paste0(path_phenos,"/Pheno_",c,".txt"), sep = "\t", row.names = FALSE, col.names = FALSE)
}
```

- Generate prevalence table

```r
Clusters = c("C1vs.Rest","C2vs.Rest","C3vs.Rest")
N_Cases = c(length(which(data$tscore_DLPC_CMC_1vsRest==1)),
            length(which(data$tscore_DLPC_CMC_2vsRest==1)),
            length(which(data$tscore_DLPC_CMC_3vsRest==1)))
N_Controls = c(length(which(data$tscore_DLPC_CMC_1vsRest==0)),
              length(which(data$tscore_DLPC_CMC_2vsRest==0)),
              length(which(data$tscore_DLPC_CMC_3vsRest==0)))

df = data.frame(Clusters,N_Cases,N_Controls)
df$N_sum = rowSums(df[,2:3])
df$Prevalence = df$N_Cases/df$N_sum

print(df)

write.table(df,paste0(path_phenos,"/Prevalence_tscore_DLPC_CMC.txt"),sep="\t", row.names = FALSE, quote = FALSE)
```

# Manhattanplot, qqplot

```r
wdir="/no-backup/lianyun/20201023_MDD_GenePathwayAssociation_Lucia/1_LifetimeMDD/11_tscore_DLPC_CMC"
setwd(wdir)

library(haplo.stats)
library(qqman)

path_gwas = paste0(wdir,"/2_GWASResults_tscore_DLPC_CMC/")
```

```r
## pre-set the join function
multi_join <- function(list_of_loaded_data, join_func, ...){
    require("dplyr")
    output <- Reduce(function(x, y) {join_func(x, y, ...)}, list_of_loaded_data)
    return(output)
}
```

```r
## readin maf file
maf = read.table("/no-backup/lianyun/biobank/ldsc/sumstatfile/Allele2_and_MAF.txt", header=TRUE)
```

```r
# read in gwas results, add allele2 and maf results, filter out low MAF and multiple basepairs, format the final summary statistics dataset, plot Manhattan, QQ, write the sumstats.
phenos=c(1,2,3) # 3 clusters
format_finals = NULL

for (p in phenos){
	gwasfilename = paste0(".","tscore_DLPC_CMC_",p,"vsRest.covararraypcs.assoc.logistic")
	print(paste0("reading in pheno_",p,"..."))

	data_files = NULL
	
	for (c in seq(1,22)){
		data = read.table(paste0(path_gwas,"chr",c,gwasfilename), header = TRUE)
		data_files[[c]] = data
		print(paste0("reading in pheno_",p,"_chromosome_",c,":done"))
	}
	print(paste0("joining chromosomes for pheno_",p))
	dataset = multi_join(data_files, full_join)
	print("joining done")

	print(paste0("checking whether ",p," matches MAF ..."))

	print(paste0("merge data pheno_",p,"with MAF ..."))
	format = data.frame(CHR = dataset$CHR, SNP = dataset$SNP, BP = dataset$BP, A1 = dataset$A1, A2 = maf$A2, MAF = maf$MAF, STAT = dataset$STAT, OR = dataset$OR, SE = dataset$SE, P = dataset$P, NMISS = dataset$NMISS)
	print("merging done")
	
	## omit NAs
	print("omitting NAs...")
	format_na = na.omit(format)
	print(" ")
	print(p)
	print(table(format_na$CHR))
	print("omitting done")

	## omit MAF<5%, omit multiple basepairs, form the new dataset into data.frame
	print("omitting MAF<5%...")
	format_hfreq = format_na[which(format_na$MAF >= 0.05),]
	#print("selecting unique alleles...")
	#format_uniqueallele = format_hfreq %>% group_by(SNP) %>% filter(dplyr::n_distinct(A1) == 1 & dplyr::n_distinct(A2) == 1)
	print("forming the dataframe...")
	format_final = data.frame(format_hfreq)

	nam = paste0("format_",p)
	assign(nam, format_final)

	format_finals[[p]] = format_final

	## Manhattan plot
	print(paste0("Plotting Manhattan_",p,"..."))
	png(paste0(wdir,"/4_ManhattanQQ_tscore_DLPC_CMC/Manhattan_",p,"_tscore_DLPC_CMC.png"))
	manhattan(format_final)
	dev.off()
	print("Manhattan done")

	## QQ plot
	print(paste0("Plotting QQ_",p,"..."))
	png(paste0(wdir,"/4_ManhattanQQ_tscore_DLPC_CMC/QQ_",p,"_tscore_DLPC_CMC.png"))
	qq(format_final$P)
	dev.off()
	print("QQ done")

	## writing sumstats
	print(paste0("writing sumstats_",p,"..."))
	write.table(format_final, paste0(wdir,"/3_GWASSumStats_tscore_DLPC_CMC/GWASSumStats_tscore_DLPC_CMC_cluster_",p), row.names = FALSE, quote = FALSE)
	print("writing done")
	print(" ")
}
```
